```python
import numpy as np
import cv2 as cv
from matplotlib import pyplot as plt
```

  # Reading in the Files in


```python
face_cascade = cv.CascadeClassifier('haarcascade_frontalface_default.xml')
eye_cascade = cv.CascadeClassifier('haarcascade_eye.xml')
img = cv.imread('silas.jpg')
gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
```

# Trying to find the face

```python
faces = face_cascade.detectMultiScale(gray, 1.3, 5)
    
for (x,y,w,h) in faces:
    cv.rectangle(img,(x,y),(x+w,y+h),(255,0,0),2)
    roi_gray = gray[y:y+h, x:x+w]
    roi_color = img[y:y+h, x:x+w]
        
plt.imshow(img[:,:,::-1])
plt.title('Silas Face')
plt.show()
```

# Trying to find the eyes

```python

img = cv.imread('silas.jpg')
gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
```

```python
faces = face_cascade.detectMultiScale(gray, 1.3, 5)
for (x,y,w,h) in faces:
   # cv.rectangle(img,(x,y),(x+w,y+h),(255,0,0),2)
    roi_gray = gray[y:y+h, x:x+w]
    roi_color = img[y:y+h, x:x+w]
    eyes = eye_cascade.detectMultiScale(roi_gray)
    for (ex,ey,ew,eh) in eyes:
        cv.rectangle(roi_color,(ex,ey),(ex+ew,ey+eh),(0,255,0),2)
        
plt.imshow(img[:,:,::-1])
plt.title('Silas Eyes')
plt.show()
```

#  Lets find the smile

```python
smile_cascade = cv.CascadeClassifier('haarcascade_smile.xml')
img = cv.imread('silas.jpg')
gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
```

```python
smiles = smile_cascade.detectMultiScale(gray, 1.3, 9)
for (x,y,w,h) in smiles:
    cv.rectangle(img,(x,y),(x+w,y+h),(255,0,0),2)
    roi_gray = gray[y:y+h, x:x+w]
    roi_color = img[y:y+h, x:x+w]
    
plt.imshow(img[:,:,::-1])
plt.title('Silas Smile')
plt.show()
```

# Combining the Facial, Eye, and Smile Recognition

```python
faces = face_cascade.detectMultiScale(gray, 1.3, 5)
    
for (x,y,w,h) in faces:
    cv.rectangle(img,(x,y),(x+w,y+h),(255,0,0),2)
    roi_gray = gray[y:y+h, x:x+w]
    roi_color = img[y:y+h, x:x+w]
    eyes = eye_cascade.detectMultiScale(roi_gray)
    for (ex,ey,ew,eh) in eyes:
        cv.rectangle(roi_color,(ex,ey),(ex+ew,ey+eh),(0,255,0),2)
        smiles = smile_cascade.detectMultiScale(gray, 1.3, 9)
        for (sx,sy,sw,sh) in smiles:
                cv.rectangle(img,(sx,sy),(sx+sw,sy+sh),(0,0,255),2)
        
plt.imshow(img[:,:,::-1])
plt.title('Silas')
plt.show()
       
```

# Running Facial and Eye recognition on multiple faces

```python
img = cv.imread('bunchfaces_2.jpg')
gray = cv.cvtColor(img,cv.COLOR_BGR2GRAY)
```

```python
faces = face_cascade.detectMultiScale(gray,1.05,25)
for (x,y,w,h) in faces:
    cv.rectangle(img,(x,y),(x+w,y+h),(255,0,0),2)
    roi_gray = gray[y:y+h, x:x+w]
    roi_color = img[y:y+h, x:x+w]
    eyes = eye_cascade.detectMultiScale(roi_gray)
    for (ex,ey,ew,eh) in eyes:
        cv.rectangle(roi_color,(ex,ey),(ex+ew,ey+eh),(0,255,0),2)

plt.imshow(img[:,:,::-1])
plt.title('Multiple Faces and Eyes')
```

# Finding peoples full bodies

```python

body_cascade = cv.CascadeClassifier('haarcascade_fullbody.xml')
img = cv.imread('fullbody.jpg')
gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
```

```python
humans = body_cascade.detectMultiScale(gray, 1.05, 1)

for (x,y,w,h) in humans:
    cv.rectangle(img,(x,y),(x+w,y+h),(200,255,20),2)
        
plt.imshow(img[:,:,::-1])
plt.title('Full Body Detection')
plt.show()
```
